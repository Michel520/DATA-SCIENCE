Dado que precio sigue siendo objeto, se debe buscar otra manera de convvertir a entero

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

Con el tratamiento aún la columna precio sigue siendo objeto, Por lo tanto se va a convertir con la funcion astype

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```
